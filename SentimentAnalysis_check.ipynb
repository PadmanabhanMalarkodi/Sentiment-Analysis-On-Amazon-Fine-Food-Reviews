## **VADER**

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
review_text = "This oatmeal is not good. It's mushy and soft, I don't like it."
sentiment = sia.polarity_scores(review_text)
sentiment
```

```python
print(sentiment['neg'])
print(sentiment['neu'])
print(sentiment['pos'])
print(sentiment['compound'])
```

```python
# Run the polarity score on the entire data frame (Text-column)
df1 = df.copy()
df1['Negative_Score'] = df['Text'].apply(lambda x: sia.polarity_scores(x)['neg'])
df1['Neutral_Score'] = df['Text'].apply(lambda x: sia.polarity_scores(x)['neu'])
df1['Positive_Score'] = df['Text'].apply(lambda x: sia.polarity_scores(x)['pos'])
df1['Compound_Score'] = df['Text'].apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax

# Initialize tokenizer and model
MODEL = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)

def polarity_scores_roberta(example):
    try:
        # Tokenize and encode the text with truncation and padding
        encoded_text = tokenizer(example, return_tensors='pt', truncation=True, padding=True, max_length=512)

        # Get model output
        with torch.no_grad():  # No need to compute gradients
            output = model(**encoded_text)

        # Get the logits and apply softmax
        scores = output[0][0].detach().numpy()
        scores = softmax(scores)

        # Create a dictionary with scores
        scores_dict = {
            'roberta_neg': scores[0],
            'roberta_neu': scores[1],
            'roberta_pos': scores[2]
        }

    except Exception as e:
        print(f"Error processing text: {example}\n{e}")
        # Return a dictionary with NaN or default values in case of an error
        scores_dict = {
            'roberta_neg': float('nan'),
            'roberta_neu': float('nan'),
            'roberta_pos': float('nan')
        }

    return scores_dict

```

```python
df1_sentiment = df1['Text'].apply(polarity_scores_roberta)
```
